```python
A = np.array(
    [[0.18, 0.53],
     [0.75, 0.47],
     [0.35, 0.21]])

B = np.array(
    [[0.17, 0.58, 0.75],
     [0.38, 0.11, 0.15]])

C = np.array([[0], [1]])

D = np.array([[0,1], [1,0]])

display(A, B, C, D)
```

```python
result_matrix = B.dot(A)
print('result_matrix: ', result_matrix, 'shape: ', result_matrix.shape)
```

```python
mult_c_matrix = np.dot(result_matrix, C)
print('mult_c_matrix: ', mult_c_matrix, 'shape: ', mult_c_matrix.shape)
```

```python
mult_c_matrix2 = result_matrix.dot(C)
print('mult_c_matrix2: ', mult_c_matrix2, 'shape: ', mult_c_matrix2.shape)
```

```python
mulyt_d_matrix = np.dot(result_matrix, D)
print('mulyt_d_matrix: ', mulyt_d_matrix, 'shape: ', mulyt_d_matrix.shape)
```
