1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



Похідна для f(x) = 10x - 5 дорівнює 10.

>тому що похідна від 10х = 10 * 1
>
>а похідна 5 = 0

Цей приклад залишила для себе, щоб нагадати собі, при прегляді теми.


```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = sp.symbols('x')
f = 10*x - 5

df = sp.diff(f, x)
print(df)
```

```python
x = sp.symbols('x')
f = 10*x - 5
df = sp.diff(f, x)

f_func = sp.lambdify(x, f)
df_value = int(df)

x_values = np.linspace(0, 10, 100)
y_values = f_func(x_values)
dy_values = np.full_like(x_values, df_value)

plt.plot(x_values, y_values, label='f(x)')
plt.plot(x_values, dy_values, label='df(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

Функція f = 10*x - 5 є лінійною, тому оцінка зростає рівномірно зі збільшенням кількості годин навчання. Похідна дорівнює 10 вона стала, отже кожна додаткова година навчання підвищює оцінку на 10 балів у всьому діапазоні 0<= x <= 10.

```python
x = sp.Symbol('x')
f = x**2 - 5*x + 1
df = sp.diff(f, x)

print(df)
```

```python
f_func = sp.lambdify(x, f)
df_func = sp.lambdify(x, df)

x_values = np.linspace(0, 100, 400)
y_values = f_func(x_values)
dy_values = df_func(x_values)

plt.plot(x_values, y_values, label='f(x^2 -5x + 1)')
plt.plot(x_values, dy_values, label='df(2*x - 5)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
